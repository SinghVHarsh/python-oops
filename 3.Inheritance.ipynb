# Inheritence

```python
class Polygon:#this is a parent or a base class
    #so here we will not be using any init method because we will not be creating any object for this class ,we will just use it as a base or a parent class
    __width=None
    __height=None
    def set_value(self,width,height):
        self.__width= width
        self.__height= height

    def get_width(self):
        return self.__width
    
    def get_height(self):
        return self.__height

class Square(Polygon):#this is a child or derived class
    def area(self):
        return self.get_height( ) * self.get_width()

class Triangle(Polygon):
    def area(self):
        return self.get_height()*self.get_width()*0.5
```

```python
s1=Square()
s1.set_value(8,15)
print(s1.area())
```

```python
t1=Triangle()
t1.set_value(5,10)
print(t1.area())
```

# understanding super function

```python
class Parent:
    def __init__(self):
        print("Parent __init__ ")
class Child(Parent):
    def __init__(self):
        print("Child __init__")
ob1=Child()
```

so as we can see here the child class init function is being initiated as we create the object but if we want it to initiate the parent class init function there are two methods that we can use

Method 1:(Without super function)

```python
class Parent:
    def __init__(self,name):
        print("Parent __init__ ",name)
class Child(Parent):
    def __init__(self):
        print("Child __init__")
        Parent.__init__(self,"Harsh")
ob1=Child()
```

so we see we are able to invoke the parent init method also

Mehtod 2:(with super function)

```python
class Parent:
    def __init__(self,name):
        print("Parent __init__ ",name)
class Child(Parent):
    def __init__(self):
        print("Child __init__")
        super().__init__("Harsh")
ob1=Child()
```

MRO(Method Resolution Order)
so if we print MRO for our child

```python
print(Child.__mro__)
```

so as we can see in the MRO first we are calling the child class then the parent class and then the object.

```python
class Parent1:
    def __init__(self,name):
        print("Parent1 __init__ ",name)

class Parent2:
    def __init__(self,marks):
        print("Parent2 __init__ ",marks)

class Child(Parent2,Parent1):
    def __init__(self):
        print("Child __init__")
        super().__init__("Harsh")
        
ob1=Child()
```

whichever class is coming first in the inheritance in child function if there are two parents then that parent init function will be called using super function if you want to access each parent init function use the first method

```python
class Parent1:
    def __init__(self,name):
        print("Parent1 __init__ ",name)

class Parent2:
    def __init__(self,marks):
        print("Parent2 __init__ ",marks)

class Child(Parent2,Parent1):
    def __init__(self):
        print("Child __init__")
        Parent1.__init__(self,"Harsh")
        Parent2.__init__(self,56)

        
ob1=Child()
```
