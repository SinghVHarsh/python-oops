# Aggregation

```python
class Salary:
    def __init__(self,pay,reward):
        self.pay1=pay
        self.reward1=reward
    def annual_salary(self):
        return (self.pay1*12)+(self.reward1)

class Employee:
    def __init__(self,name,position,pay,reward):
        self.name1=name
        self.position1=position
        self.final_salary=Salary(pay,reward)
    
    def final_salary_m(self):
        return self.final_salary.annual_salary()
```

```python
e1=Employee("Harsh","DataEngineer",100000,100000)
e1.final_salary_m()
```

# Abstract Classes

```python
class Shape():
    def area(self):
        pass
    def perimeter(self):
        pass
class Square(Shape):
    def __init__(self,side):
        self.__side=side
sh=Shape()
sq=Square(20)
```

```python
from abc import ABC,abstractmethod
class Shape(ABC):#declaring ABC inside my class makes it as an abstract class means no one can create an instance of my parent class.
    @abstractmethod#this makes mandatory for an abstract method to be defined inside a child class
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
class Square(Shape):
    def __init__(self,side):
        self.__side=side
sh=Shape()
sq=Square(20)
```

So as we can see we are not able to instance of my parent class Shape because it is an abstract class.

```python
from abc import ABC,abstractmethod
class Shape(ABC):#declaring ABC inside my class makes it as an abstract class means no one can create an instance of my parent class.
    @abstractmethod#this makes mandatory for an abstract method to be defined inside a child class
    def area(self):
        pass
    @abstractmethod#this is a decorator
    def perimeter(self):
        pass
class Square(Shape):
    def __init__(self,side):
        self.__side=side
    def area(self):
        return self.__side*self.__side
    def perimeter(self):
        return self.__side*4

sq=Square(20)
print(sq.area())
```

# Operator Overloading



```python



class Books:
    def __init__(self,pages):
        self.pages=pages

b1=Books(100)
b2=Books(150)
print(b1+b2)
```

so as we can see we cant add the content of two of the instances directly since the plus operator has been defined for adding two integers string values ,to add the content of the instance directly we have to use operator overloading.


```python
class Books:
    def __init__(self,pages):
        self.pages=pages
    def __add__(self,other):
        return self.pages+other.pages
        

b1=Books(100)
b2=Books(150)
print(b1+b2)
```

lets add operator overloading for our multiplication function

```python
class Books:
    def __init__(self,pages):
        self.pages=pages
    def __add__(self,other):
        return self.pages+other.pages
    def __mul__(self,other):
        return self.pages*other.pages
        

b1=Books(100)
b2=Books(150)
print(b1+b2)
print(b1*b2)
```

so as we can see here the code stops while running so its an runtime error,it has sucessfully compiled but while running it threw an error of Zero Division error.Because of that after our result variable the code stops working.This can be a dangerous situation when working with complex websites .So we will use exception handling for this.

```python
result=None
x=int(input("Enter the first number: "))
y=int(input("Enter the second number: "))

try:
    result=x/y #so on this line we want to handle our exception therefore we will add our try block before this line
except:
    print("Error with our code")

print("---New Line---")
print("Result is ",result)
```

so as we can see we gave x as 10 and y as 0 so it should give us an zero division error but we handles that error by an exception and the whole programme ran successfully.

# Try except else and finally

```python
# so if we want to add the reason of the error also then 
result=None
x=int(input("Enter the first number: "))
y=int(input("Enter the second number: "))

try:
    result=x/y #so on this line we want to handle our exception therefore we will add our try block before this line
except Exception as e:
    print(e)
else:
    print("Inside else")
finally:
    print("Inside finally")

print("---New Line---")
print("Result is ",result)
```

so if there is any exception we will touch try except and finally
and if there is no error or exception we will touch try else and finally

# Raising Exception


```python
class Tea:
    def __init__(self,temperature):
        self.__temperature=temperature
    def drink_tea(self):
        if self.__temperature>85:
            raise ValueError("Tea too Hot")
        elif self.__temperature<65:
            print("Cold to drink")
        else:
            print('Tea OK to drink')

cup=Tea(100)
cup.drink_tea()
```

# creating user defined exceptions


```python
class TeaHotException(Exception):
    def __init__(self,arg):
        self.msg=arg

class TeaColdException(Exception):
    def __init__(self,arg):
        self.msg=arg

class Tea:
    def __init__(self,temperature):
        self.__temperature=temperature
    def drink_tea(self):
        if self.__temperature>85:
            raise TeaHotException("Tea too Hot")
        elif self.__temperature<65:
            raise TeaColdException("Cold to drink")
        else:
            print('Tea OK to drink')

cup=Tea(100)
cup.drink_tea()
```
